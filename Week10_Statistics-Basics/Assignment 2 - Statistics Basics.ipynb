### Q1. What are the three measures of central tendency?

The three measures of central tendency are:

1. **Mean**: The arithmetic average of a dataset.
2. **Median**: The middle value when the dataset is ordered from smallest to largest.
3. **Mode**: The value that appears most frequently in the dataset.

### Q2. What is the difference between the mean, median, and mode? How are they used to measure the central tendency of a dataset?

- **Mean**:
  - **Definition**: Sum of all values divided by the number of values.
  - **Usage**: Provides a measure of the average value in a dataset, useful when data is symmetrically distributed without outliers.

- **Median**:
  - **Definition**: The middle value in an ordered dataset.
  - **Usage**: Provides the central value in a dataset, especially useful for skewed data or when there are outliers.

- **Mode**:
  - **Definition**: The most frequently occurring value in a dataset.
  - **Usage**: Identifies the most common value, useful for categorical data or when you want to know the most frequent observation.

### Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe the spread of a dataset? Provide an example.

- **Range**: Difference between the maximum and minimum values. Shows the total spread.
  - Example: For data [5, 10, 15], Range = 15 - 5 = 10

- **Variance**: Average of squared deviations from the mean. Measures the average squared deviation from the mean.
  - Example: For data [5, 10, 15], Mean = 10, Variance = [(5-10)² + (10-10)² + (15-10)²] / 3 = 50 / 3 ≈ 16.67

- **Standard Deviation**: Square root of variance. Provides a measure of spread in the same units as the data.
  - Example: For data [5, 10, 15], Standard Deviation = √16.67 ≈ 4.08

### Q8. What do you understand about skewness in data?

- **Definition**: Skewness measures the asymmetry of the distribution of data. It indicates whether data is skewed to the left (negative skewness) or right (positive skewness).

### Q9. If a data is right skewed then what will be the position of median with respect to mean?

- **Right Skewed Data**: The mean is typically greater than the median. This is because the mean is pulled towards the tail on the right.

### Q10. Explain the difference between covariance and correlation. How are these measures used in statistical analysis?

- **Covariance**:
  - **Definition**: A measure of how two variables change together. Positive covariance indicates that variables increase together, while negative covariance indicates that one increases as the other decreases.
  - **Usage**: Shows the direction of the linear relationship but not the strength or consistency.

- **Correlation**:
  - **Definition**: A standardized measure of the relationship between two variables, ranging from -1 to 1. Positive correlation indicates a direct relationship, while negative correlation indicates an inverse relationship.
  - **Usage**: Indicates both the strength and direction of the linear relationship.

### Q11. What is the formula for calculating the sample mean? Provide an example calculation for a dataset.

- **Formula**: \(\bar{x} = \frac{\sum x_i}{n}\)
  - **Example**: For dataset [5, 7, 9]:
    - Mean = (5 + 7 + 9) / 3 = 21 / 3 = 7

### Q12. For a normal distribution data, what is the relationship between its measure of central tendency?

- **Normal Distribution**: The mean, median, and mode are all equal and located at the center of the distribution.

### Q13. How is covariance different from correlation?

- **Covariance**: Measures the directional relationship between two variables but is affected by the scale of the variables.
- **Correlation**: Measures both the strength and direction of the linear relationship between two variables and is normalized to be independent of the scale of the variables.

### Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.

- **Mean**: Outliers can significantly affect the mean, pulling it towards the outlier value.
  - **Example**: For data [1, 2, 3, 4, 100], the mean is 22, which is skewed by the outlier 100.

- **Median**: Less affected by outliers, as it depends only on the middle value.
  - **Example**: For the same data [1, 2, 3, 4, 100], the median is 3.

- **Variance and Standard Deviation**: Outliers can increase the variance and standard deviation, indicating more spread in the data.
  - **Example**: For data [1, 2, 3, 4, 100], the standard deviation is much larger compared to a similar dataset without the outlier.

These measures help describe different aspects of the data and understand how individual data points impact the overall summary statistics.
